## 2. [Inference] Posterior Probability Inference: Belief Propagation & Variable Elimination.

The goal is to compute the posterior distribution of $X_{q}$ such that $X_{q} \in \{S, A, B, L, T, E\}$.
$$ 
\begin{align}
\mathbb{P}(\; X_{q} \;\vert \; F = 1, X = 0)\ .
\end{align}
$$

- In order to understand Belief Propagation (BP), and thus derive the necessary equations for implementation in code, the best way is to think from the perspective of each Random Variable (RV) $x$ in the network.
- Each RV $x$ in the network can get influenced by evidence variables ${\sf Evid}$ either from it's parents or from it's children. Accordingly, pseudo-evidence variables ${\sf Evid^{+}}, {\sf Evid^{-}}$ denote evidence ${\sf Evid}$ accessible through parents and childeren of RV $x$, respectievly. Mathematically, this can be understood as:
$$ 
\begin{align}
    \mathbb{P}(\; x \;\vert \; {\sf Evid} ) = \frac{\mathbb{P}(x, {\sf Evid})}{\mathbb{P}({\sf Evid})} = \frac{\mathbb{P}(x, {\sf Evid^{+}}, {\sf Evid^{-}})}{\mathbb{P}({\sf Evid^{+}}, {\sf Evid^{-}})}   
    \propto \; & {\sf constant} \cdot \mathbb{P}(\; x \;\vert \; {\sf Evid^{+}} ) \cdot \mathbb{P}(\; {\sf Evid^{-}} \;\vert \; x)\ , \\
    \propto \; & {\sf constant} \cdot \pi(x) \cdot \lambda(x)\ , \\
    \propto \; & {\sf Belief}(x) 
\end{align}
$$

- What BP essentially does is that it computes the intermedite factors (or messages) $\pi(x)$ and $\lambda(x)$ that are common to many elimination orderings.
- Next, we will do just that, i.e, initialize these messages that are to be propagated back and forth in the bayesian network in order to compute these beliefs for each RV $x \in \mathcal{G}$.

```python
def conv_to_index(assignments):
    """
    Returns an array of indices associated with each assignment in assignments.
    
    input:   assignments: np.array():
    output:  index:       np.array():
    """
    index = []
    rows, cols = assignments.shape
    for row in range(rows):
        temp = 0
        for col in range(cols):
            if assignments[row, col]==0:
                temp += (1) * (2 ** (col))
            else:
                temp += (0) * (2 ** (col))
        index.append(temp + 1)
    return np.array(index)

def dec_to_bin(number, num_bits):
    """
    Returns an array of binary assignments that each RV realizes. 
    """
    answer = []
    for charac in f'{number:0{num_bits}b}'[::-1]:
        if charac=='0':
            answer.append(1)
        elif charac=='1':
            answer.append(0)

    return np.array(answer)

def conv_to_assignment(valueList, length_of_assignment):
    """
    Returns assignments in binary form associated with each value in valueList.
    
    input:  valueList: List:            A list of numbers for which binary assignments are to be generated.
            length_of_assignment: int:  Determines the length of the binary number.
     
    output: assignments: np.array():    An assignment table associated with each index of the desired .val representation
    """
    assignments = np.empty(shape=[0, length_of_assignment], dtype=np.int32)
    for element in valueList:
        temp = dec_to_bin(element, length_of_assignment)
        assignments = np.vstack([assignments, temp])
    return assignments

def product_factor(factor1, factor2):  
    """
    Computes the product of two factors (fundamental computation to solve many downstream tasks).
    
    input:  factor1: Struct 
            factor2: Struct
           
    output: factor3: Struct: another factor that computes their product.
    """
    # determining variables in the new factor.
    factor3 = Struct()
    factor3.var = sorted( list( set(factor1.var + factor2.var) ) ) # a sorted non-repeated union
    #print("factor3 variables: ", factor3.var)
    
    # determining the cardinality of the new factor.
    _, mapfactor1 = ismember(factor1.var, factor3.var)
    _, mapfactor2 = ismember(factor2.var, factor3.var)
    factor3.card = [0] * len(factor3.var)
    #print("mapfactor1: ", mapfactor1) # factor 1 me jo elements hai vo factor 3 mei kaha milengey?
    #print("mapfactor2: ", mapfactor2)
    
    for index, element in enumerate(mapfactor1):
        factor3.card[element] = factor1.card[index]
    for index, element in enumerate(mapfactor2):
        factor3.card[element] = factor2.card[index]
    
    # determining the probability values associated with the new factor.
    totalNumberofValues = math.prod(factor3.card)
    factor3.val = [-1] * totalNumberofValues
    assignments = conv_to_assignment( list( range(totalNumberofValues) ),  int(math.log(totalNumberofValues)/math.log(2)) ) 
    
    indxfactor1 = conv_to_index( assignments[:, mapfactor1] );  # acessing factor1.val elements
    indxfactor2 = conv_to_index( assignments[:, mapfactor2] );  # acessing factor2.val elements
    
    for index in range(totalNumberofValues):
        factor3.val[index] = factor1.val[ indxfactor1[index]-1 ] * factor2.val[ indxfactor2[index]-1 ]
        
    return factor3

def sum_marginalization(factor1, variableOut):
    """
    Marginalizes out variables in variableOut from factor1.
    
    input:  factor1: Struct  
            variableOut: List: A list of variables to be marginalized out from factor1.var
            
    output: Struct
    """
    # determining variables in the new factor.
    factor3 = Struct()
    factor3.var = sorted( list( set(factor1.var) - set(variableOut) ) )
    
    # determining the cardinality of the new factor.
    mapfactor3 = []
    for randomVar in factor3.var:
        for index in range(len(factor1.var)):
            if factor1.var[index] == randomVar:
                mapfactor3.append(index)
                break
    factor3.card = [0] * len(factor3.var)
    
    for idx in range(len(factor3.card)):
        factor3.card[idx] = factor1.card[ mapfactor3[idx] ]
    
    # yeh pata karo ki factor3 me jo variables hai vo factor1 me kaha milengey.
    #_, mapfactor1 = ismember(factor1.var, factor3.var)
    #for idx in mapB:
    #    B.card.append(A.card[idx])
    
    totalNumberofValuesfactor1 = math.prod(factor1.card)
    assignments = conv_to_assignment( list( range(totalNumberofValuesfactor1) ), int(math.log(totalNumberofValuesfactor1)/math.log(2)) )
    indxfactor3 = conv_to_index( assignments[:, mapfactor3] );
   
    # determining the probability values associated with the new factor.
    totalNumberofValuesfactor3 = math.prod(factor3.card)
    factor3.val = [0] * totalNumberofValuesfactor3
    
    for i in range(1, totalNumberofValuesfactor3+1):
        for j in range(totalNumberofValuesfactor1):
            if indxfactor3[j]==i:
                factor3.val[i-1] = factor3.val[i-1] + factor1.val[j]
                
    return factor3

def normalization(probaValues):
    answer = [-1, -1]
    answer[0] = probaValues[0]/(sum(probaValues))
    answer[1] = probaValues[1]/(sum(probaValues))
    return answer
```

### 2.1 Inference Algorithm-1: Belief Propagation

```python
# Belief Propagation algorithm initialization 

# PI and LAMBDA messages for RVs without parents.
PI_A = Struct([let_to_num['A']], [2], [0.6, 0.4])
LAMBDA_A = Struct([let_to_num['A']], [2], [1, 1])

PI_S = Struct([let_to_num['S']], [2], [0.8, 0.2])
LAMBDA_S = Struct([let_to_num['S']], [2], [1, 1])

PI_L = Struct([let_to_num['L']], [2], [0.8, 0.2])
LAMBDA_L = Struct([let_to_num['L']], [2], [1, 1])

# PI and LAMBDA messages for RVs that are observed as evidence.
PI_X = Struct([let_to_num['X']], [2], [0, 1])
LAMBDA_X = Struct([let_to_num['X']], [2], [0, 1])

PI_F = Struct([let_to_num['F']], [2], [1, 0])
LAMBDA_F = Struct([let_to_num['F']], [2], [1, 0])
```

```python
# Implementation of the Belief Propagation algorithm.

for _ in range(3):
    # Forming PI_T
    temp = product_factor(probab_T, PI_A)
    #print(temp.var, temp.card, temp.val)
    PI_T = sum_marginalization(temp, [let_to_num['A']])
    #print(PI_T.var, PI_T.card, PI_T.val)

    # Forming PI_E
    temp = product_factor(probab_E, product_factor(PI_T, PI_L))
    #print(temp.var, temp.card, temp.val)
    PI_E = sum_marginalization(temp, [let_to_num['L'], let_to_num['T']])
    #print(PI_E.var, PI_E.card, PI_E.val)

    # Forming PI_B
    temp = product_factor(probab_B, PI_S)
    #print(temp.var, temp.card, temp.val)
    PI_B = sum_marginalization(temp, [let_to_num['S']])
    #print(PI_B.var, PI_B.card, PI_B.val)

    # Computing LAMBDA(B)
    temp = product_factor(probab_X, LAMBDA_X)
    LAMBDA_X_TO_E = sum_marginalization(temp, [let_to_num['X']])
    PI_E_TO_F = product_factor(PI_E, LAMBDA_X_TO_E)

    temp = product_factor(probab_F, PI_E_TO_F)
    temp_1 = sum_marginalization(temp, [let_to_num['E']])
    temp_2 = product_factor(temp_1, LAMBDA_F)
    LAMBDA_B = sum_marginalization(temp_2, [let_to_num['F']])

    # Computing LAMBDA(E)
    temp = product_factor(probab_F, PI_B)
    temp_1 = sum_marginalization(temp, [let_to_num['B']])
    temp_2 = product_factor(temp_1, LAMBDA_F)
    LAMBDA_F_TO_E = sum_marginalization(temp_2, [let_to_num['F']])
    LAMBDA_E = product_factor(LAMBDA_F_TO_E, LAMBDA_X_TO_E)

    # Computing LAMBDA(T)
    temp = product_factor(probab_E, PI_L)
    temp_1 = sum_marginalization(temp, [let_to_num['L']])
    temp_2 = product_factor(temp_1, LAMBDA_E)
    LAMBDA_E_TO_T = sum_marginalization(temp_2, [let_to_num['E']])
    LAMBDA_T = LAMBDA_E_TO_T

    # Computing LAMBDA(S)
    temp = product_factor(probab_B, LAMBDA_B)
    LAMBDA_B_TO_S = sum_marginalization(temp, [let_to_num['B']])
    LAMBDA_S = LAMBDA_B_TO_S

    # Computing LAMBDA(L)
    temp = product_factor(probab_E, PI_T)
    temp_1 = sum_marginalization(temp, [let_to_num['T']])
    temp_2 = product_factor(temp_1, LAMBDA_E)
    LAMBDA_T_TO_E = sum_marginalization(temp_2, [let_to_num['E']])
    LAMBDA_L = LAMBDA_T_TO_E

    # Computing LAMBDA(A)
    temp = product_factor(probab_T, LAMBDA_T)
    LAMBDA_T_TO_A = sum_marginalization(temp, [let_to_num['T']])
    LAMBDA_A = LAMBDA_T_TO_A
    #print(" ")
```

### 2.2 Inference Algorithm-2: Variable Elimination

```python
def compute_joint_distribution(probab_tables):
    """
    Computes the joint distribution given a list of probability factors as input.
    
    input:   probab_tables: List of Struct's
    
    output:  joint:         Struct
    """
    joint = probab_tables[0]
    for index, probab_table in enumerate(probab_tables):
        if index >0:
            joint = product_factor(joint, probab_table)
            
    return joint

def observe_evidence(probab_tables, evidence):
    """
    Incorporates evidence in the 
    
    input:  probab_tables: List of Struct's
            evidence:      np.array([])
    
    output: probab_tables: List of Struct's: the function modifies the input factor to incorporate evidence.
    
    """
    for rowIndex, element in enumerate(evidence):
        #print(rowIndex, element)
        variable = element[0] # which RV is evidence?
        value    = element[1] # the value it takes?
        #print(variable, value)

        for probab_table in probab_tables:
            if variable in probab_table.var:
                totalNumberofValues = math.prod(probab_table.card)
                assignments = conv_to_assignment( list( range(totalNumberofValues) ),  int(math.log(totalNumberofValues)/math.log(2)) )
                #print(assignments)
                for ii, rowAssignment in enumerate(assignments):
                    for jj, randVar in enumerate(probab_table.var):
                        if randVar==variable:
                            if assignments[ii, jj]!=value:
                                length = assignments[ii, :].shape[0]
                                index_to_be_assigned = conv_to_index(assignments[ii, :].reshape((1, length)))
                                index_to_be_assigned = index_to_be_assigned.item()
                                probab_table.val[ index_to_be_assigned-1 ] = 0
                       
    return probab_tables

def compute_marginal(variables_in_marginal, probab_tables, evidence):
    """
    input:  variables_in_marginal: list of variables that you want in the marginal
                    probab_tables: List of Struct's
                         evidence: np.array([])
    
    output:        marginalFactor: Struct
    
    """
    probab_tables_with_evidence = observe_evidence(probab_tables, evidence)
    joint_with_evidence = compute_joint_distribution(probab_tables_with_evidence)
    
    variableOut = sorted( list( set(joint_with_evidence.var) - set(variables_in_marginal) ) )
    marginalFactor = sum_marginalization(joint_with_evidence, variableOut)
    #print(marginalFactor.var, marginalFactor.card, marginalFactor.val)
    
    marginalFactor.val = list(np.array(marginalFactor.val)/sum(marginalFactor.val))
    return marginalFactor
```

## 3. [Learning] Parameter Learning of Each CPD:

### 3.1 Generating a Dataset: Ancestral Sampling

- Here, we will have to employ other techniques since the above is not possible for Discrete BNs. 
- I follow the most obvious choice for datset generation Ancestral sampling.

```python
numSamples = 1000
data = []

for _ in range(numSamples):
    sample = []
    
    # intializing the probability tables
    probab_A = Struct([let_to_num['A']], [2], [0.6, 0.4])
    probab_S = Struct([let_to_num['S']], [2], [0.8, 0.2])
    probab_L = Struct([let_to_num['L']], [2], [0.8, 0.2])
    probab_T = Struct([let_to_num['T'], let_to_num['A']], [2, 2], [0.8, 0.2, 0.3, 0.7])
    probab_E = Struct([let_to_num['E'], let_to_num['L'], let_to_num['T']], [2, 2, 2], [0.8, 0.2, 0.7, 0.3, 0.6, 0.4, 0.1, 0.9])
    probab_X = Struct([let_to_num['X'], let_to_num['E']], [2, 2], [0.8, 0.2, 0.1, 0.9])
    probab_B = Struct([let_to_num['B'], let_to_num['S']], [2, 2], [0.7, 0.3, 0.3, 0.7])
    probab_F = Struct([let_to_num['F'], let_to_num['B'], let_to_num['E']], [2, 2, 2], [0.9, 0.1, 0.8, 0.2, 0.7, 0.3, 0.2, 0.8])
    probab_tables = [probab_A, probab_S, probab_L, probab_T, probab_E, probab_X, probab_B, probab_F]

    # sampling A
    probaValue = probab_tables[0].val
    realization_A = np.random.choice(2, 1, p=probaValue[::-1])
    #print("A: ", realization_A)
    sample.append(realization_A.item())

    # sampling S
    probaValue = probab_tables[1].val
    realization_S = np.random.choice(2, 1, p=probaValue[::-1])
    #print("S: ", realization_S)
    sample.append(realization_S.item())

    # sampling L
    probaValue = probab_tables[2].val
    realization_L = np.random.choice(2, 1, p=probaValue[::-1])
    #print("L: ", realization_L)
    sample.append(realization_L.item())

    # sampling T
    evidence = np.array( [[1, realization_A.item()]] )
    originalFactor = probab_tables[3]
    originalFactorListWithEvidence = observe_evidence([originalFactor], evidence)
    #print(originalFactorListWithEvidence[0].var, originalFactorListWithEvidence[0].card, originalFactorListWithEvidence[0].val)
    marginalizedFactor = sum_marginalization(originalFactorListWithEvidence[0], [1])
    #print(marginalizedFactor.var, marginalizedFactor.card, marginalizedFactor.val)
    probaValue = marginalizedFactor.val
    realization_T = np.random.choice(2, 1, p=probaValue[::-1])
    #print("T: ", realization_T)
    sample.append(realization_T.item())

    # sampling E
    evidence = np.array( [[4, realization_T.item()], 
                          [3, realization_L.item()]] )
    originalFactor = probab_tables[4]
    originalFactorListWithEvidence = observe_evidence([originalFactor], evidence)
    #print(originalFactorListWithEvidence[0].var, originalFactorListWithEvidence[0].card, originalFactorListWithEvidence[0].val)
    marginalizedFactor = sum_marginalization(originalFactorListWithEvidence[0], [3, 4])
    #print(marginalizedFactor.var, marginalizedFactor.card, marginalizedFactor.val)
    probaValue = marginalizedFactor.val
    realization_E = np.random.choice(2, 1, p=probaValue[::-1])
    #print("E: ", realization_E)
    sample.append(realization_E.item())

    # sampling X
    evidence = np.array( [[5, realization_E.item()]] )
    originalFactor = probab_tables[5]
    #print(originalFactor.var, originalFactor.card, originalFactor.val)
    originalFactorListWithEvidence = observe_evidence([originalFactor], evidence)
    #print(originalFactorListWithEvidence[0].var, originalFactorListWithEvidence[0].card, originalFactorListWithEvidence[0].val)
    marginalizedFactor = sum_marginalization(originalFactorListWithEvidence[0], [5])
    #print(marginalizedFactor.var, marginalizedFactor.card, marginalizedFactor.val)
    probaValue = marginalizedFactor.val
    realization_X = np.random.choice(2, 1, p=probaValue[::-1])
    #print("X: ", realization_X)
    sample.append(realization_X.item())

    # sampling B
    evidence = np.array( [[2, realization_S.item()]] )
    originalFactor = probab_tables[6]
    #print(originalFactor.var, originalFactor.card, originalFactor.val)
    originalFactorListWithEvidence = observe_evidence([originalFactor], evidence)
    #print(originalFactorListWithEvidence[0].var, originalFactorListWithEvidence[0].card, originalFactorListWithEvidence[0].val)
    marginalizedFactor = sum_marginalization(originalFactorListWithEvidence[0], [2])
    #print(marginalizedFactor.var, marginalizedFactor.card, marginalizedFactor.val)
    probaValue = marginalizedFactor.val
    realization_B = np.random.choice(2, 1, p=probaValue[::-1])
    #print("B: ", realization_B)
    sample.append(realization_B.item())

    # sampling F
    evidence = np.array( [[7, realization_B.item()],
                          [5, realization_E.item()]] )
    originalFactor = probab_tables[7]
    #print(originalFactor.var, originalFactor.card, originalFactor.val)
    originalFactorListWithEvidence = observe_evidence([originalFactor], evidence)
    #print(originalFactorListWithEvidence[0].var, originalFactorListWithEvidence[0].card, originalFactorListWithEvidence[0].val)
    marginalizedFactor = sum_marginalization(originalFactorListWithEvidence[0], [5, 7])
    #print(marginalizedFactor.var, marginalizedFactor.card, marginalizedFactor.val)
    probaValue = marginalizedFactor.val
    realization_F = np.random.choice(2, 1, p=probaValue[::-1])
    #print("F: ", realization_F)
    sample.append(realization_F.item())
    
    data.append(sample)
    #print(sample)
    #print(" ")

print(" ")
data = np.array(data)
print(data.shape)
print(data)
```
